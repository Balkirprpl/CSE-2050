```python
import dateparser
import datetime

def checkSchedule(date, s, e, schedule):
    # looping through every range in the schedule
    for i in range(len(schedule)):
        sections_str = schedule[i].split('-')
        startTime = int(dateparser.parse(sections_str[0]).strftime('%H%M'))
        endTime = int(dateparser.parse(sections_str[1]).strftime('%H%M'))
        
        # setting both ranges of time 
        range1 = range(startTime, endTime)
        range2 = range (s,e)
        
        # checking if the ranges overlap
        if s in range1 or e in range1 or startTime in range2 or endTime in range2:
            print(schedule[i], "overlaps with", date) # outputting
            return True
    return False


# testing the script with the sample input
print(" Sample output:")

# initializing lists
time_periods = ["6:30am-7:00am", "8:00am-9:00am",
"6:00am-7:30am", "7:00pm-9:00pm", "8:00pm-8:30pm"]
schedule = []

for date in time_periods:
    sections_str = date.split('-') # splitting the times into two HH:MM parts
    
    # parsing both times
    startTime = dateparser.parse(sections_str[0]).strftime('%H%M') 
    endTime = dateparser.parse(sections_str[1]).strftime('%H%M')
    
    # checking for collisions and adding to the list of schedules
    if not checkSchedule(date, int(startTime), int(endTime), schedule):
        schedule.append(date)


# testing the script with the given input
print("\n\n Office hours 7:45am - 9:00am:")

# initializing lists
requested_times = ["7:45am-8:10am", "7:50am-8:00am",
"8:00am-8:19am", "8:15am-8:30am", "8:45am-9:00am",
"8:30am-8:35am", "8:30am-8:40am"]
schedule2 = []

for date in requested_times:
    sections_str = date.split('-') # splitting the times into two HH:MM parts
    
    # parsing both times
    startTime = dateparser.parse(sections_str[0]).strftime('%H%M')
    endTime = dateparser.parse(sections_str[1]).strftime('%H%M')
    
    # checking for collisions and adding to the list of schedules
    if not checkSchedule(date, int(startTime), int(endTime), schedule2):
        schedule2.append(date)
    
```

```python
def printHeader():
    print('["Country", "Gold", "Silver", "Bronze", "Total"]')

def printList(lst):
    for i in range(0, len(lst)):
        print(lst[i])
            
def weighted_sum(data: list, rank_vector: list):
    total = 0
    index = 0
    for i in range (1, 4):
        if str(data[i]).isnumeric():
            total += rank_vector[index] * data[i]
        index += 1    
    return total

def sort_by_key(data, weight):
    for a in data:
        a[4] = weighted_sum(a, weight)
    return sorted(data, key=lambda x: x[4], reverse=True) # sorting the list by the total weighted value

def delete_dash(data):
    for a in data:
        for i in range(1,4):
            if str(a[i]).isnumeric():
                continue
            else:
                a[i] = 0
 
medals = [
["Country", "Gold", "Silver", "Bronze", "Total"],
["Australia", 1, 2, 1, 4], # sum = 8
["Austria", 7, 7, 4, 18], # sum = 39
["Belarus", "-", 2, "-", 2], # sum = 4
["Belgium", 1, "-", 1, 2], # sum = 4
["Canada", 4, 8, 14, 26], # sum = 42
["Czech Republic", 1, "-", 1, 2], # sum = 4
["Estonia", "-", "-", 1, 1], # sum = 1
["Finland", 2, 2, 4, 8], # sum = 14
["France", 5, 7, 2, 14], # sum = 31
["Germany", 12, 10, 5, 27],# sum = 61
["Great Britain", 1, 1, "-", 2], # sum = 5
["Hungary", 1, "-", 2, 3], # sum = 5
["Italy", 2, 7, 8, 17], # sum = 28
["Japan", 3, 6, 9, 18], # 
["Latvia", "-", "-", 1, 1],
["Netherlands", 8, 5, 4, 17],
["New Zealand", 2, 1, "-", 3],
["Norway", 16, 8, 13, 37],
["People's Republic of China", 9, 4, 2, 15],
["Poland", "-", "-", 1, 1],
["ROC", 6, 12, 14, 32],
["Republic of Korea", 2, 5, 2, 9],
["Slovakia", 1, "-", 1, 2],
["Slovenia", 2, 3, 2, 7],
["Spain", "-", 1, "-", 1],
["Sweden", 8, 5, 5, 18],
["Switzerland", 7, 2, 5, 14],
["Ukraine", "-", 1, "-", 1],
["United States of America", 8, 10, 7, 25],
]


# removing unnecesarry sections of the table
delete_dash(medals)
del medals[0]

# printing original table
print("Original table:")
printHeader()
printList(medals)

# sorting by Gold-Silver-Bronze
print("\n\n")
print("Table sorted by Gold-Silver-Bronze:")
medals_sorted = sorted(medals, key=lambda x: (x[1], x[2], x[3]), reverse=True)
printList(medals_sorted)

# sorting by Fibonacci weighted point system 
print("\n\n")
print("Table sorted by Fibonnacci weighted point system:")
medals_sorted = sort_by_key(medals, [3, 2, 1])
printHeader()
printList(medals_sorted)

# sorting by Exponential weighted point system 
print("\n\n")
print("Table sorted by Exponential weight point system:")
medals_sorted = sort_by_key(medals, [4, 2, 1])
printHeader()
printList(medals_sorted)

# sorting by Topend Sports weighted point system
print("\n\n")
print("Table sorted by Topen Sports weighted point system:")
medals_sorted = sort_by_key(medals, [6, 2, 1])
printHeader()
printList(medals_sorted)
```
